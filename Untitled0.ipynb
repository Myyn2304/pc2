```python
import pandas as pd
link='https://github.com/Myyn2304/pc2/raw/refs/heads/main/HDI.xlsx'
table=pd.read_excel(link, header=4)
table
```

```python
#las columnas de mi tabla, de las cuales solo me importan algunos
table.columns
```

```python
#quitar el espacio de los nombres cambiandolo por _ no sin antes str.strip() quita los espacios al
#inicio y final de cada nombre para que así no queden cosas como
#'Human_Development_Index_(HDI)_' sino 'Human_Development_Index_(HDI)'
table.columns=table.columns.str.strip().str.replace(r'\s+', '_', regex=True)
table.columns
```

```python
#eliminar las columnas que no sirven y/o que no son necesarias y las dos primeras filas
toDrop=['Unnamed:_0','Unnamed:_3','Unnamed:_5','Unnamed:_7','Unnamed:_9','Unnamed:_11','Unnamed:_13','GNI_per_capita_rank_minus_HDI_rank', 'HDI_rank']
table.drop(columns=toDrop,inplace=True)
table.drop(index=[0,1], inplace=True)
```

```python
#cambiar nombres a los requeridos
nombre2={'Unnamed:_1':'Country','Human_Development_Index_(HDI)':'Human_Development_Index','Gross_national_income_(GNI)_per_capita':'Gross_national_income_per_capita'}
table.rename(columns=nombre2,inplace=True)
table.columns
```

```python
table
```

```python
#comprobar si hay valores nan
table[table.isnull().any(axis=1)]
```

```python
#como si hay filas con nan, procedemos a borrarlas
table=table.dropna(axis='rows',how='any',ignore_index=True)
table
```

```python
#eliminando todas las filas que no sean países (las filas de abajo de Somalia)
table = table.drop(index=range(195, 210))
table
```

```python
#qué tipo de dato es cada columna
table.info()
```

```python
# se observa que el unico problema son los ".." de los ultimos 4 paises
pd.unique(table.Human_Development_Index[~table.Human_Development_Index.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# en esta columna todos los valores encajan con valores numéricos
pd.unique(table.Life_expectancy_at_birth[~table.Life_expectancy_at_birth.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# se observa que el unico problema son los ".." de los ultimos 4 paises
pd.unique(table.Expected_years_of_schooling[~table.Expected_years_of_schooling.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# se observa que el unico problema son los ".." de los ultimos 4 paises
pd.unique(table.Mean_years_of_schooling[~table.Mean_years_of_schooling.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
# se observa que el unico problema son los ".." de los ultimos 4 paises
pd.unique(table.Gross_national_income_per_capita[~table.Gross_national_income_per_capita.astype(str).str.contains(r'\d+\.\d+', regex=True)])
```

```python
#transformar todos los valores a numeros en las tablas, se espera que los NaN esten en los ultimos 4 paises
table.Human_Development_Index=pd.to_numeric(table.Human_Development_Index,errors='coerce')
table.Life_expectancy_at_birth=pd.to_numeric(table.Life_expectancy_at_birth,errors='coerce')
table.Expected_years_of_schooling=pd.to_numeric(table.Expected_years_of_schooling,errors='coerce')
table.Mean_years_of_schooling=pd.to_numeric(table.Mean_years_of_schooling,errors='coerce')
table.Gross_national_income_per_capita=pd.to_numeric(table.Gross_national_income_per_capita,errors='coerce')
table
```

```python
#se observa que ahora todos son numeros
table.info()
```

```python
#creando la nueva categoria para clasificarlos en base a HDI
table['Category']=pd.cut(table.Human_Development_Index,
                           bins=[0,0.549,0.699,0.796,0.962],
                           labels=['LOW HUMAN DEVELOPMENT','MEDIUM HUMAN DEVELOPMENT','HIGH HUMAN DEVELOPMENT','VERY HIGH HUMAN DEVELOPMENT'],
                           ordered=True)
table
```

```python
#por si algún país no tiene mayúscula
table['Country']=table.Country.str.title()
table
```
